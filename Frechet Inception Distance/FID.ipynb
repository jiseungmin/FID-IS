```python
import json
import matplotlib.pyplot as plt

insertDataList = []

with open("./metric-fid50k_full.jsonl", 'r') as f:
    for line in f:
        data = json.loads(line)
        insertDataList.append(data.get("results").get("fid50k_full"))

plt.plot(insertDataList,linewidth=2,color="Blue",label="FID Over Time",marker="o")
plt.xlim(0,58)
plt.xticks(range(0, 58, 5))
plt.legend(loc='upper right')
plt.title("StyleGan2-ADA FID")
plt.xlabel('Time(h)')
plt.ylabel("FID")
plt.savefig("FID_Image.png")
plt.show()
```
